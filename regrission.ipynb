```python

sell_cols = [
    'cat3_apartment-sell',
    'cat3_house-villa-sell',
    'cat3_industry-agriculture-business-sell',
    'cat3_office-sell',
    'cat3_shop-sell'
]


rent_cols = [
    'cat3_house-villa-rent',
    'cat3_industry-agriculture-business-rent',
    'cat3_office-rent',
    'cat3_shop-rent'
]
def get_target_value(row):
    if any(row[col] == 1 for col in sell_cols):
        return row['price_value']
    elif any(row[col] == 1 for col in rent_cols):
        return row['rent_value']
    else:
        return None


divar['target_value'] = divar.apply(get_target_value, axis=1)

divar = divar.dropna(subset=['target_value'])


divar['target_value'] = divar['target_value'].astype(float)
```
