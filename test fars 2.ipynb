گزارش تحلیل تفاوت میانگین مساحت ساختمان‌ها بین کلان‌شهرها و شهرهای کوچک
مقدمه
در این تحلیل، هدف بررسی این فرضیه است که میانگین مساحت ساختمان‌های مسکونی در کلان‌شهرها نسبت به شهرهای کوچک تفاوت دارد. به عبارت دیگر، می‌خواهیم ببینیم آیا اندازه ساختمان‌ها در این دو دسته از شهرها به طور معناداری متفاوت است یا خیر.

فرضیات آزمون
فرض صفر (H0): میانگین مساحت ساختمان‌ها در کلان‌شهرها برابر با میانگین مساحت ساختمان‌ها در شهرهای کوچک است.
(یعنی تفاوت معناداری بین دو گروه وجود ندارد.)

فرض مقابل (H1): میانگین مساحت ساختمان‌ها در کلان‌شهرها با میانگین مساحت ساختمان‌ها در شهرهای کوچک متفاوت است.
(یعنی تفاوت معناداری بین دو گروه وجود دارد.)

روش تحلیل
بررسی بصری توزیع داده‌ها:
نمودار هیستوگرام اندازه ساختمان‌ها برای هر دو گروه کلان‌شهرها و شهرهای کوچک رسم شد تا شکل کلی توزیع مشاهده شود.

تست نرمال بودن:
برای بررسی نرمال بودن داده‌ها، تست Shapiro-Wilk بر روی داده‌های هر گروه انجام شد. نتایج نشان داد که توزیع اندازه ساختمان‌ها در هر دو گروه از نرمال بودن تبعیت نمی‌کند (p-value بسیار کمتر از 0.05).

انتخاب آزمون آماری مناسب:
با توجه به عدم نرمال بودن داده‌ها، از آزمون غیرپارامتری Mann-Whitney U برای مقایسه توزیع اندازه ساختمان‌ها بین دو گروه استفاده شد.


نتایج آزمون Mann-Whitney U
آماره آزمون (U)	مقدار p-value
98799351216.0	0.0

مقدار p-value بسیار کوچک (کمتر از 0.05) است که نشان‌دهنده رد فرض صفر است.

نتیجه‌گیری
با توجه به نتایج به دست آمده، تفاوت معناداری در اندازه ساختمان‌ها بین کلان‌شهرها و شهرهای کوچک وجود دارد. بنابراین، فرضیه تحقیق تایید می‌شود که میانگین مساحت ساختمان‌ها در این دو دسته شهرها متفاوت است.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import shapiro, ttest_ind
divar_t=divar.copy()
# فرض: دیتافریم divar_t بارگذاری شده است

# 1. حذف ردیف‌هایی که construction_year مقدارشون خالیه
divar_t = divar_t.dropna(subset=['construction_year', 'building_size'])

# 2. نگه داشتن فقط مقادیر عددی در construction_year
mask = divar_t['construction_year'].astype(str).str.isnumeric()
divar_t = divar_t[mask].copy()

# 3. تبدیل construction_year به عدد صحیح
divar_t['construction_year'] = divar_t['construction_year'].astype(int)

# 4. حذف ردیف‌هایی که building_size مقدارشون خالی یا غیر عددیه
divar_t['building_size'] = pd.to_numeric(divar_t['building_size'], errors='coerce')
divar_t = divar_t.dropna(subset=['building_size'])

# 5. جداکردن خانه‌های قدیمی (قبل از ۱۳۹۶) و جدید (از ۱۳۹۶ به بعد)
old_homes = divar_t[divar_t['construction_year'] < 1396]['building_size']
new_homes = divar_t[divar_t['construction_year'] >= 1396]['building_size']

# --- رسم هیستوگرام اولیه ---
plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.hist(old_homes, bins=50, alpha=0.7, label='old')
plt.hist(new_homes, bins=50, alpha=0.7, label='new')
plt.title('hist')
plt.xlabel('masahat')
plt.ylabel('tedad nemone')
plt.legend()

# 6. نرمال‌سازی با تبدیل لگاریتمی (log)
# چون ممکنه صفر یا منفی باشه، قبلش مقدار min رو چک می‌کنیم و در صورت لزوم جابجایی انجام می‌دیم
min_size = min(old_homes.min(), new_homes.min())
if min_size <= 0:
    offset = abs(min_size) + 1
else:
    offset = 0

old_homes_log = np.log(old_homes + offset)
new_homes_log = np.log(new_homes + offset)

# --- رسم هیستوگرام بعد از نرمال‌سازی ---
plt.subplot(1,2,2)
plt.hist(old_homes_log, bins=50, alpha=0.7, label='old(log)')
plt.hist(new_homes_log, bins=50, alpha=0.7, label='new(log)')
plt.title('hist')
plt.xlabel('masahat')
plt.ylabel('tedad nemone')
plt.legend()

plt.tight_layout()
plt.show()

# 7. آزمون شاپیرو برای نرمال بودن (روی داده‌های نرمال شده)
stat_old, p_old = shapiro(old_homes_log.sample(5000))  # نمونه ۵۰۰۰ تایی برای شاپیرو (برای سرعت)
stat_new, p_new = shapiro(new_homes_log.sample(5000))

print(f"Shapiro Test - خانه‌های قدیمی: statistic={stat_old}, p-value={p_old}")
print(f"Shapiro Test - خانه‌های جدید: statistic={stat_new}, p-value={p_new}")

# 8. آزمون تی مستقل (t-test)
t_stat, p_value = ttest_ind(old_homes_log, new_homes_log, equal_var=False)
print(f"t-statistic: {t_stat}")
print(f"p-value: {p_value}")
```

```python
from scipy.stats import mannwhitneyu

old_homes = divar_t[divar_t['construction_year'] < 1396]['building_size'].dropna()
new_homes = divar_t[divar_t['construction_year'] >= 1396]['building_size'].dropna()

u_stat, p_value = mannwhitneyu(old_homes, new_homes, alternative='two-sided')

print(f"Mann-Whitney U statistic: {u_stat}")
print(f"p-value: {p_value}")
```

گزارش آزمون فرض آماری: بررسی تفاوت میانگین مساحت خانه‌های قدیمی و جدید
فرضیه مسئله:
معمولاً گفته می‌شود که «خانه‌های قدیمی دلبازتر و بزرگ‌تر هستند». برای بررسی این فرضیه، میانگین مساحت خانه‌های قدیمی (ساخت قبل از سال ۱۳۹۶) با خانه‌های جدید (ساخت از سال ۱۳۹۶ به بعد) مقایسه می‌شود.

فرضیات آزمون:

فرض صفر (H₀): میانگین مساحت خانه‌های قدیمی و جدید برابر است.

فرض مقابل (H₁): میانگین مساحت خانه‌های قدیمی بیشتر از خانه‌های جدید است.

روش انجام آزمون:

ابتدا داده‌ها از نظر سال ساخت دسته‌بندی شدند.

داده‌های دارای مقدار معتبر سال ساخت جدا شدند و به دو گروه خانه‌های قدیمی و جدید تقسیم شدند.

آزمون نرمال بودن داده‌ها با آزمون Shapiro-Wilk انجام شد که نشان داد داده‌ها نرمال نیستند.

بنابراین از آزمون ناپارامتری Mann-Whitney U برای مقایسه دو گروه استفاده شد.

همچنین آزمون t-test با فرض نابرابری واریانس‌ها (Welch’s t-test) نیز محاسبه شد.

نتایج آزمون Shapiro-Wilk:

خانه‌های قدیمی: p-value بسیار کوچک (کمتر از ۰.۰۰۰۱) – داده‌ها نرمال نیستند.

خانه‌های جدید: p-value بسیار کوچک (کمتر از ۰.۰۰۰۱) – داده‌ها نرمال نیستند.

نتایج آزمون Mann-Whitney U:

آماره U = 57339561683.5

p-value = 0.0 (کمتر از ۰.۰۵)
این نتیجه نشان می‌دهد که تفاوت معناداری بین دو گروه وجود دارد.

نتیجه‌گیری:
با توجه به p-value بسیار کوچک، فرض صفر رد شده و می‌توان با اطمینان آماری گفت که میانگین مساحت خانه‌های قدیمی به طور معناداری کمتر از خانه‌های جدید است.
بنابراین، برخلاف تصور عمومی، داده‌ها نشان می‌دهند که خانه‌های جدید مساحت بیشتری دارند و فرضیه اولیه‌ی «خانه‌های قدیمی دلبازتر و بزرگ‌تر هستند» تأیید نمی‌شود.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind, mannwhitneyu, shapiro
from sklearn.preprocessing import PowerTransformer

# آماده‌سازی داده
df_test = divar.copy()
df = df_test.dropna(subset=['price_value'])

# تعریف امکانات لاکچری و غیر لاکچری
luxury_features = ['has_pool', 'has_barbecue', 'has_sauna', 'has_jacuzzi']
non_luxury_features = ['has_balcony', 'has_elevator', 'has_parking', 'has_warehouse']

# ایجاد ستون 'has_luxury'
df['has_luxury'] = df[luxury_features].any(axis=1)
```

```python
# استخراج قیمت‌ها برای گروه‌های لاکچری و غیرلاکچری
luxury_yes = df[df['has_luxury']]['price_value']
luxury_no = df[~df['has_luxury']]['price_value']
```

```python


# Shapiro برای نرمال بودن
n_yes = min(5000, len(luxury_yes))
n_no = min(5000, len(luxury_no))

shapiro_lux_yes = shapiro(luxury_yes.sample(n_yes, random_state=1))
shapiro_lux_no = shapiro(luxury_no.sample(n_no, random_state=1))

print(f"Shapiro Test - دارای امکانات لاکچری: statistic={shapiro_lux_yes.statistic}, p-value={shapiro_lux_yes.pvalue}")
print(f"Shapiro Test - فاقد امکانات لاکچری: statistic={shapiro_lux_no.statistic}, p-value={shapiro_lux_no.pvalue}")
```

```python
# رسم نمودار توزیع قبل از نرمال‌سازی
plt.figure(figsize=(12, 5))
sns.kdeplot(luxury_yes, label='emkanat lactury')
sns.kdeplot(luxury_no, label='faghed emkanat lactury')
plt.title('nemodar tozie geymat')
plt.xlabel('price')
plt.legend()
plt.show()
```

```python
pt = PowerTransformer(method='yeo-johnson')
combined_data = pd.concat([luxury_yes, luxury_no], axis=0).values.reshape(-1, 1)
normalized_data = pt.fit_transform(combined_data).flatten()

normalized_yes = normalized_data[:len(luxury_yes)]
normalized_no = normalized_data[len(luxury_yes):]
```

```python
plt.figure(figsize=(12, 5))
sns.kdeplot(normalized_yes, label='lactury')
sns.kdeplot(normalized_no, label='geir lactury')
plt.title("nemodar tozie geymat")
plt.xlabel("price")
plt.legend()
plt.show()
```

```python
u_stat, p_val = mannwhitneyu(luxury_yes, luxury_no, alternative='two-sided')
print(f"Mann-Whitney U statistic: {u_stat}")
print(f"p-value: {p_val}")
```

گزارش آزمون فرض آماری: بررسی تاثیر امکانات لاکچری بر قیمت ملک
فرضیه‌ی مسئله:

در بسیاری از آگهی‌های ملکی، وجود امکانات لاکچری مانند استخر، باربیکیو، سونا و جکوزی به‌عنوان مزیتی برای افزایش ارزش ملک ذکر می‌شود. هدف این تحلیل، بررسی این موضوع است که آیا وجود این امکانات واقعاً تأثیر معناداری بر میانگین قیمت ملک دارد یا خیر.

فرضیات آزمون:
فرض صفر (H₀): میانگین قیمت املاک دارای امکانات لاکچری و فاقد آن برابر است.

فرض مقابل (H₁): میانگین قیمت املاک دارای امکانات لاکچری با فاقد آن متفاوت است.

روش آزمون:
داده‌ها از نظر وجود امکانات لاکچری (استخر، باربیکیو، سونا، جکوزی) به دو گروه تقسیم شدند.

توزیع داده‌ها با استفاده از آزمون Shapiro-Wilk بررسی شد:

نتایج نشان داد که داده‌ها نرمال نیستند (p-value ≪ 0.05).

با توجه به غیرنرمال بودن توزیع، از آزمون ناپارامتری Mann-Whitney U استفاده شد.

نتایج:
آماره آزمون: 1,598,070,461.0

p-value: 0.0017 (کمتر از 0.05)

نتیجه‌گیری:
با توجه به اینکه مقدار p-value بسیار کمتر از 0.05 است، فرض صفر رد می‌شود. بنابراین، می‌توان با اطمینان آماری گفت که وجود امکانات لاکچری باعث تفاوت معنادار در قیمت ملک می‌شود.

این نتیجه نشان می‌دهد که امکانات لاکچری مانند استخر، باربیکیو، سونا و جکوزی می‌توانند به‌طور چشم‌گیری بر قیمت فروش ملک تأثیرگذار باشند.
